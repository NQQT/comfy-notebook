```python
# This is design to get asset remotely, from URL and so on

import os
from urllib.parse import urlsplit

from IPython.display import clear_output


def extract_filename(url_link):
    url_path = urlsplit(url_link)
    return os.path.basename(url_path.path)


# For Fetching Asset Remotely
def asset_fetch(asset_type, asset_url, asset_name=None):
    # Go to the correct folder
    %cd {working_dir} / ComfyUI / models / {asset_type}

    # Setting the asset name as requied
    if not asset_name is None:
        # Set it to asset name
        file_name = asset_name
    else:
        file_name = extract_filename(asset_url)

    # for loading files
    if not os.path.exists(file_name):
        os.system(f'wget -O "{file_name}" "{asset_url}"')

    # Return to the working directory
    %cd {working_dir}
    # Clearing the output to show python has been updated
    clear_output()


# Getting Custom Node from Github Repo (or similar)
def git_custom_node(git_url, commit=None):
    # Reading the Folder Name
    folder_name = extract_filename(git_url)

    %cd {working_dir} / ComfyUI / custom_nodes
    # Remove whatever version it is
    !rm -rf {folder_name}
    !git clone {git_url}

    # Go into the folder
    %cd {working_dir} / ComfyUI / custom_nodes / {folder_name}

    # Do we need checkout to a specific version?
    if not commit is None:
        !git checkout {commit}

    # Install the requirement (if it exists)
    if os.path.exists("requirements.txt"):
        !{pip} install -r requirements.txt

    # Return to the working directory
    %cd {working_dir}

    # Clearing the output to show python has been updated
    # clear_output()
```
